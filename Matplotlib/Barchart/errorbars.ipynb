```python
from loggerfile import Log
import matplotlib.pyplot as plt
import numpy as np
```

```python
def errorbar():
    '''
    Description : program to create bar plots with error bars on the same figure.
    '''
    try:
        mean=[0.2474, 0.1235, 0.1737, 0.1824]
        deviation=[0.3314, 0.2278, 0.2836, 0.2645]
        index=np.arange(len(mean))

        plt.bar(index,mean,0.35,yerr=deviation, color="blue")
        plt.xlabel("Velocity")
        plt.ylabel("Score")
        plt.title("Bar plots with error bars")
        plt.minorticks_on()
        plt.show()
    except Exception as e:
        loggerfile.Log.logging.error(e)   
```

```python
errorbar()
```
