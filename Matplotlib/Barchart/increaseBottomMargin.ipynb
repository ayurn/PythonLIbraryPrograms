```python
from matplotlib import pyplot as plt
from loggerfile import Log
```

```python
def incmargin():
    '''
    Description : function to increase margin
    '''
    try:
        languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
        popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

        plt.bar(languages,popularity, edgecolor='black')
        plt.xlabel('Languages')
        plt.ylabel('Popularity')
        plt.title('Languages with popularity')
        plt.subplots_adjust(bottom=0.4)
        plt.minorticks_on()
        plt.grid(which='major',linestyle='-',color='blue')
        plt.grid(which='minor',linestyle=":",color='black')
        plt.show()
    except Exception as e:
        loggerfile.Log.logging.error(e)   
```

```python
incmargin()
```
