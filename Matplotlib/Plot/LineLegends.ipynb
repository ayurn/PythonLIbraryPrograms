```python
from matplotlib import pyplot as plt
from loggerfile import Log
```

```python
def legends():
    '''
    Description : function to make line legends.
    '''
    try:
        plt.plot([1,2,3,4,5],[9,8,7,6,5],label="First Line")
        plt.plot([1,2,3,4,5],[4,3,2,1,1],label="Second Line")
        plt.plot([1,2,3,4,5],[0,0,9,0,0],label="Third Line")
        plt.xlabel("y-axis")
        plt.ylabel("x-axis")
        plt.title("Display Legends")
        plt.legend()
        plt.show()
    except Exception as e:
        loggerfile.Log.logging.error(e)
```

```python
legends()
```
