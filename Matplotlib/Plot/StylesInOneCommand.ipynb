```python
def diffstyles():
    '''
    Description : plot several lines with different format styles in one command using arrays.
    '''
    try:
        x=np.array([1,2,3,4,5])
        plt.plot(x,x,"--",x,x**2,"-.",x,x**3,':')
        plt.show()
    except Exception as e:
        loggerfile.Log.logging.error(e)    
```
