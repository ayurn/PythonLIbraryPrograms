```python
from matplotlib import pyplot as plt
from loggerfile import Log
```

```python
def useraxis():
    
    '''
    Description: function to draw a line using given axis values with suitable label
    '''
    try:

        x=list(map(int,input("Enter x-axis values : ").split()))
        y=list(map(int,input("Enter y-axis values : ").split()))

        plt.plot(x,y)
        plt.xlabel("X-axis")
        plt.ylabel("Y-axis")
        plt.title("Line (User axis)")
        plt.show()
    except Exception as e:
        loggerfile.Log.logging.error(e)
```
