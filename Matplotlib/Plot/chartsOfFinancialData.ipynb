```python
from matplotlib import pyplot as plt
import pandas as pd 
from loggerfile import Log
```

```python
def financialChart():
    '''
    Description : function to draw line charts of the financial data of Alphabet Inc
    '''
    try:
        df=pd.read_csv("fdata.csv")
        df.set_index('Date',inplace=True)
        plt.plot(df)
        plt.xlabel("Dates")
        plt.ylabel("Income")
        plt.title("Financial Data of Alphabet Inc.")
        plt.legend(df.columns)
        plt.show()
    except Exception as e:
        loggerfile.Log.logging.error(e)

            
```

```python
financialChart()
```
