```python
from matplotlib import pyplot as plt
from matplotlib import style
from loggerfile import Log
```

```python
def styles():
    '''
    Description : creating different styles of plot
    '''
    try:
        style.use("ggplot")
        x=[1,2,3,4,5]
        plt.plot(x,[9,8,7,6,5],'r',label="First Line",linewidth=5)
        plt.xticks(x)
        plt.xlabel("y-axis")
        plt.ylabel("x-axis")
        plt.title("Display Legends")
        plt.show()

        style.use("grayscale")
        plt.plot([1,2,3,4,5],[4,3,2,1,1],'c',label="Second Line",linewidth=8)
        plt.xlabel("y-axis")
        plt.ylabel("x-axis")
        plt.title("Display Legends")
        plt.show()
        
        style.use("dark_background")
        plt.plot([1,2,3,4,5],[0,0,9,0,0],'g',label="Third Line",linewidth=3)
        plt.xlabel("y-axis")
        plt.ylabel("x-axis")

        plt.title("Display Legends")
        plt.legend()
        plt.show()
    except Exception as e:
        loggerfile.Log.logging.error(e)
```

```python
styles()
```
