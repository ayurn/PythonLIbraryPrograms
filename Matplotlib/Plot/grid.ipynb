```python
from matplotlib import pyplot as plt
import pandas as pd
from loggerfile import Log
```

```python
def grids():
    try:
        dataFrame=pd.read_csv("fidata.csv")
        dataFrame.set_index('Date',inplace=True)
        plt.plot(dataFrame,'b')
        plt.grid(color='c',ls='-',lw=.5)
        plt.title("Closing value of Alphabet lnc.")
        plt.xlabel("x-axis")
        plt.show()
    except Exception as e:
        loggerfile.Log.logging.error(e)    
```

```python
grids()
```
