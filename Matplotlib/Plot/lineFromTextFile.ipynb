```python
from matplotlib import pyplot as plt
import pandas as pd
from loggerfile import Log
```

```python
def linefromtxt():
    '''
    Description : function to draw a line using text file.
    '''
    try:
        df=pd.read_table("test.txt",delim_whitespace=True,names=('x','y'))
        df.set_index('x',inplace=True)
        plt.plot(df)
        plt.xlabel("X-axis")
        plt.ylabel("Y-axis")
        plt.title("Line (using text file)")
        plt.show()
    except Exception as e:
        loggerfile.Log.logging.error(e)
```
