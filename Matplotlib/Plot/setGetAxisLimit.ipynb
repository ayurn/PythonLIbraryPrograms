```python
from matplotlib import pyplot as plt
from matplotlib import style
from loggerfile import Log
```

```python
def limits():
    '''
    Description : display the current axis limits values and set new axis values.
    '''
    try:
        x=[1,2,3,4,5]
        plt.plot(x,[9,8,7,6,5],"-.")
        axis=plt.gca()
        print("Before setting axis limits:",axis.get_xlim(),axis.get_ylim())
        axis.set_xlim([0,5])
        axis.set_ylim([4,9])
        print("After setting axis limits:",axis.get_xlim(),axis.get_ylim())
    except Exception as e:
        loggerfile.Log.logging.error(e)
```

```python
limits()
```
