```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from loggerfile import Log
```

```python
try:
    data = pd.read_csv('sea_level_data.csv')
    time = data['Time']
    sea_levels = data[['GMSL']]

    plt.scatter(time, sea_levels, edgecolors='g', facecolor = 'b')
    plt.xlabel('Time')
    plt.ylabel('Sea Level (inches)')
    plt.title('Rise in Sealevel')
    plt.show()
except Exception as e:
        loggerfile.Log.logging.error(e) 
```
